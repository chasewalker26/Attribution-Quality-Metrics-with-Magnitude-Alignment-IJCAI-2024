```python
# https://github.com/chasewalker26/Magnitude-Aligned-Scoring
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
##### The results of the sensitivity test for the metrics #####
# each file has 30 rows. 5 rows per metric. In order SIC, AIC, RISE ins, RISE del, MAS ins, MAS del. 

with open('../test_results/R101/ImageNet_constant_sensitivity_1000_images.csv', newline='') as csvfile:
# with open('../test_results/VIT16/ImageNet_constant_sensitivity_1000_images.csv', newline='') as csvfile:
# with open('../test_results/R101/resisc45_constant_sensitivity_1000_images.csv', newline='') as csvfile:
# with open('../test_results/VIT16/resisc45_constant_sensitivity_1000_images.csv', newline='') as csvfile:
# with open('../test_results/R101/ImageNet_noised_sensitivity_1000_images.csv', newline='') as csvfile:
# with open('../test_results/VIT16/ImageNet_noised_sensitivity_1000_images.csv', newline='') as csvfile:
# with open('../test_results/R101/resisc45_noised_sensitivity_1000_images.csv', newline='') as csvfile:
# with open('../test_results/VIT16/resisc45_noised_sensitivity_1000_images.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
```

```python
def norm_0_1(array):
    return (array - min(array)) / (max(array) - min(array))

from sklearn.metrics import auc

for i in range(30):
    data[i] = np.asarray(data[i], dtype=float)

SIC_AIC_x = np.linspace(0, 1, 1001)
MAS_RISE_x = np.linspace(0, 1, 225)

plt.rcParams.update({'font.size': 30})
fig, axs = plt.subplots(3, 3, figsize = (30, 22))

fig.suptitle('Noised Sensitivity: ImageNet, R101', fontsize = 45)
plt.rc('lines', linewidth=4.0)

###### Graph SIC [0 - 4] ######
axs[0, 0].plot(SIC_AIC_x, data[0], label = "Original")
axs[0, 0].plot(SIC_AIC_x, data[1], label = "Add 0.05%", ls = "dashed", color = "red")
axs[0, 0].plot(SIC_AIC_x, data[2], label = "Add 1%", ls = "dashdot", color = "green")
axs[0, 0].plot(SIC_AIC_x, data[3], label = "Add 2%",  ls = "dotted", color = "purple")
axs[0, 0].plot(SIC_AIC_x, data[4], label = "Add 3%", ls = "dashdot", color = "black")
axs[0, 0].legend(fontsize="20")
axs[0, 0].set_title("SIC Noise Recognition")
axs[0, 0].set_ylabel("Score")
axs[0, 0].set_xlabel("Perturbation Percentage")

sum = 0
for i in range(1, 5):
    sum += np.abs(data[0] - data[i])
dist = np.mean(sum) / 5

sic_auc = auc(SIC_AIC_x, data[0])
perc_diff = dist / sic_auc * 100
print("SIC Percent Diff: " + str(round(perc_diff, 3)))
print()

###### Graph AIC [5 - 9] ######
axs[1, 0].plot(SIC_AIC_x, data[5], label = "Original")
axs[1, 0].plot(SIC_AIC_x, data[6], label = "Add 0.05%", ls = "dashed", color = "red")
axs[1, 0].plot(SIC_AIC_x, data[7], label = "Add 1%", ls = "dashdot", color = "green")
axs[1, 0].plot(SIC_AIC_x, data[8], label = "Add 2%", ls = "dotted", color = "purple")
axs[1, 0].plot(SIC_AIC_x, data[9], label = "Add 3%", ls = "dashdot", color = "black")
axs[1, 0].legend(fontsize="20")
axs[1, 0].set_title("AIC Noise Recognition")
axs[1, 0].set_ylabel("Score")
axs[1, 0].set_xlabel("Perturbation Percentage")

axs[2, 0].axis("off")

sum = 0
for i in range(6, 10):
    sum += np.abs(data[5] - data[i])
dist = np.mean(sum) / 4

aic_auc = auc(SIC_AIC_x, data[5])
perc_diff = dist / aic_auc * 100
print("AIC Percent Diff: " + str(round(perc_diff, 3)))
print()

###### Graph RISE INS [6 - 8] ######
axs[0, 1].plot(MAS_RISE_x, data[10], label = "Original")
axs[0, 1].plot(MAS_RISE_x, data[11], label = "Add 0.05%", ls = "dashed", color = "red")
axs[0, 1].plot(MAS_RISE_x, data[12], label = "Add 1%", ls = "dashdot", color = "green")
axs[0, 1].plot(MAS_RISE_x, data[13], label = "Add 2%", ls = "dotted", color = "purple")
axs[0, 1].plot(MAS_RISE_x, data[14], label = "Add 3%", ls = "dashdot", color = "black")
axs[0, 1].legend(fontsize="20")
axs[0, 1].set_title("RISE Ins Noise Recognition")
axs[0, 1].set_ylabel("Score")
axs[0, 1].set_xlabel("Perturbation Percentage")

sum = 0
for i in range(11, 15):
    sum += np.abs(data[10] - data[i])
dist = np.mean(sum) / 4

ins_auc = auc(MAS_RISE_x, data[10])
perc_diff = dist / ins_auc * 100
print("RISE Ins Percent Diff: " + str(round(perc_diff, 3)))
print()

###### Graph RISE DEL [9 - 11] ######
axs[1, 1].plot(MAS_RISE_x, data[15], label = "Original")
axs[1, 1].plot(MAS_RISE_x, data[16], label = "Add 0.05%", ls = "dashed", color = "red")
axs[1, 1].plot(MAS_RISE_x, data[17], label = "Add 1%", ls = "dashdot", color = "green")
axs[1, 1].plot(MAS_RISE_x, data[18], label = "Add 2%", ls = "dotted", color = "purple")
axs[1, 1].plot(MAS_RISE_x, data[19], label = "Add 3%", ls = "dashdot", color = "black")
axs[1, 1].legend(fontsize="20")
axs[1, 1].set_title("RISE Del Noise Recognition")
axs[1, 1].set_ylabel("Score")
axs[1, 1].set_xlabel("Perturbation Percentage")

sum = 0
for i in range(16, 20):
    sum += np.abs(data[15] - data[i])
dist = np.mean(sum) / 4

del_auc = auc(MAS_RISE_x, data[15])
perc_diff = dist / (1 - del_auc) * 100
print("RISE Del Percent Diff: " + str(round(perc_diff, 3)))
print()


###### Graph RISE Diff ######
axs[2, 1].plot(MAS_RISE_x, norm_0_1(data[10] - data[15]), label = "Original")
axs[2, 1].plot(MAS_RISE_x, norm_0_1(data[11] - data[16]), label = "Add 0.05%", ls = "dashed", color = "red")
axs[2, 1].plot(MAS_RISE_x, norm_0_1(data[12] - data[17]), label = "Add 1%", ls = "dashdot", color = "green")
axs[2, 1].plot(MAS_RISE_x, norm_0_1(data[13] - data[18]), label = "Add 2%", ls = "dotted", color = "purple")
axs[2, 1].plot(MAS_RISE_x, norm_0_1(data[14] - data[19]), label = "Add 3%", ls = "dashdot", color = "black")
axs[2, 1].legend(fontsize="20")
axs[2, 1].set_title("RISE Diff Noise Recognition")
axs[2, 1].set_ylabel("Score")
axs[2, 1].set_xlabel("Perturbation Percentage")

sum = 0
for i in range(16, 20):
    sum += np.abs(norm_0_1(data[10] - data[15]) - norm_0_1(data[i - 5] - data[i]))
dist = np.mean(sum) / 4

ins_del_auc = auc(MAS_RISE_x, norm_0_1(data[10] - data[15]))
perc_diff = dist / ins_del_auc * 100
print("RISE Diff Percent Diff: " + str(round(perc_diff, 3)))
print()


###### Graph MAS INS [12 - 14] ######
axs[0, 2].plot(MAS_RISE_x, data[20], label = "Original")
axs[0, 2].plot(MAS_RISE_x, data[21], label = "Add 0.05%", ls = "dashed", color = "red")
axs[0, 2].plot(MAS_RISE_x, data[22], label = "Add 1%", ls = "dashdot", color = "green")
axs[0, 2].plot(MAS_RISE_x, data[23], label = "Add 2%", ls = "dotted", color = "purple")
axs[0, 2].plot(MAS_RISE_x, data[24], label = "Add 3%", ls = "dashdot", color = "black")
axs[0, 2].legend(fontsize="20")
axs[0, 2].set_title("MAS Ins Noise Recognition")
axs[0, 2].set_ylabel("Score")
axs[0, 2].set_xlabel("Perturbation Percentage")

sum = 0
for i in range(21, 25):
    sum += np.abs(data[20] - data[i])
dist = np.mean(sum) / 4

ins_auc = auc(MAS_RISE_x, data[20])
perc_diff = dist / ins_auc * 100
print("MAS Ins Percent Diff: " + str(round(perc_diff, 3)))
print()


###### Graph MAS DEL [15 - 17] ######
axs[1, 2].plot(MAS_RISE_x, data[25], label = "Original")
axs[1, 2].plot(MAS_RISE_x, data[26], label = "Add 0.05%", ls = "dashed", color = "red")
axs[1, 2].plot(MAS_RISE_x, data[27], label = "Add 1%", ls = "dashdot", color = "green")
axs[1, 2].plot(MAS_RISE_x, data[28], label = "Add 2%", ls = "dotted", color = "purple")
axs[1, 2].plot(MAS_RISE_x, data[29], label = "Add 3%", ls = "dashdot", color = "black")
axs[1, 2].legend(fontsize="20")
axs[1, 2].set_title("MAS Del Noise Recognition")
axs[1, 2].set_ylabel("Score")
axs[1, 2].set_xlabel("Perturbation Percentage")

sum = 0
for i in range(26, 30):
    sum += np.abs(data[25] - data[i])
dist = np.mean(sum) / 4

del_auc = auc(MAS_RISE_x, data[25])
perc_diff = dist / (1 - del_auc) * 100
print("MAS Del Percent Diff: " + str(round(perc_diff, 3)))
print()

###### Graph RISE Diff ######
axs[2, 2].plot(MAS_RISE_x, norm_0_1(data[20] - data[25]), label = "Original")
axs[2, 2].plot(MAS_RISE_x, norm_0_1(data[21] - data[26]), label = "Add 0.05%", ls = "dashed", color = "red")
axs[2, 2].plot(MAS_RISE_x, norm_0_1(data[22] - data[27]), label = "Add 1%", ls = "dashdot", color = "green")
axs[2, 2].plot(MAS_RISE_x, norm_0_1(data[23] - data[28]), label = "Add 2%", ls = "dotted", color = "purple")
axs[2, 2].plot(MAS_RISE_x, norm_0_1(data[24] - data[29]), label = "Add 3%", ls = "dashdot", color = "black")
axs[2, 2].legend(fontsize="20")
axs[2, 2].set_title("MAS Diff Noise Recognition")
axs[2, 2].set_ylabel("Score")
axs[2, 2].set_xlabel("Perturbation Percentage")

sum = 0
for i in range(26, 30):
    sum += np.abs(norm_0_1(data[20] - data[25]) - norm_0_1(data[i - 5] - data[i]))
dist = np.mean(sum) / 4

ins_del_auc = auc(MAS_RISE_x, norm_0_1(data[20] - data[25]))
perc_diff = dist / ins_del_auc * 100
print("MAS Diff Percent Diff: " + str(round(perc_diff, 3)))
print()

plt.subplots_adjust(wspace = 0.25, hspace=0.4, top = 0.92)
plt.show()
```
